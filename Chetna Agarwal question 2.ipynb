```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data2 = data[(data['unit'] == 'cases')]
data2 = data[(data['location'] == 'Argentina-Catamarca') | (data['location'] == 'Argentina-CABA')]
len(data2)
```

```python
data2.head()
```

```python
data2 = data2.reset_index()
data2.head()
```

```python
data2['value_new'] = [float(data2['value'][i]) for i in range(0,len(data2))]
data2.head()
```

```python
c1 = data2[data2['location'] == 'Argentina-CABA']
c2 = data2[data2['location'] == 'Argentina-Catamarca']
```

```python
date1 = np.array(c1['report_date'])
arr1 = np.array(c1['value_new'])
date1, arr1
```

```python
date2 = np.array(c2['report_date'])
arr2 = np.array(c2['value_new'])
date2, arr2
```

```python
date2 == date1
```

```python
plt.figure(figsize=(20,10))
plt.plot(date1, arr1, label = 'Argentina-CABA', color='blue')
plt.plot(date2, arr2, label = 'Argentina-Catamarca', color ='red')
plt.legend()
plt.show()
```
