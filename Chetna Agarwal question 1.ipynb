QUESTION 1 - Show data field code 'AR0005' values for 'Argentina-Catamarca' for entire duration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head(10)
```

```python
import pandas as pd
data = data[pd.notnull(data['report_date'])]
data['report_date'].unique()
```

```python
data2 = data[(data['location'] == 'Argentina-Catamarca') &(data['data_field_code'] == 'AR0005')]
```

```python
code = data2['data_field_code']
code_arr = np.array(code)
code.unique()
```

```python
data2
```

```python
values = data2['value']
value_arr = np.array(values)
value_arr
```

```python
report_date = data2['report_date']
date_arr = np.array(report_date)
date_arr
```

```python
dates = []
for i in range(0, len(date_arr)):
    dates = dates + re.findall(r"[\w']+", str(date_arr[i]))
```

```python
len(dates)/3
```

```python
y=[]
m=[]
d=[]
for i in range(0, len(dates)):
    if i%3 == 0:
        y.append(int(dates[i]))
    elif i%3 == 1:
        m.append(int(dates[i]))
    elif i%3 == 2:
        d.append(int(dates[i]))
```

```python
data2['year'] = y
data2['month'] = m
data2['date'] = d
data2
```

```python
type(data2['report_date'][16])
```

```python
data2['value_new'] = [int(i) for i in data2['value']]
data2.head()
```

```python
value = np.array(data2['value_new'])
month = np.array(data2['month'])
year = np.array(data2['year'])
date = np.array(data2['date'])
value, month, year, date
```

```python
import datetime
data2['datetime'] = [datetime.datetime(year[i], month[i], date[i]) for i in range(len(date))]
data2
```
